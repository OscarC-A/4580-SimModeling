We will now replicate their experiments.

**a.** Generate $n = 1000$ copies of a simple random walk $S^i = (S^i_0,S^i_1,\ldots,S^i_{T})$ (with $T=100$ and $i\in\{1,2,\ldots,n\}$). For each walk, compute $Z^i_T = \frac{S^i_T}{T}$ (i.e., the per-round winnings of Paul after $100$ games), as well as $W^i_T = \frac{1}{T}\sum_{t=1}^{T}\mathbb{1}_{\{S^i_t>0\}}$ (i.e., the fraction of rounds in $[1,T]$ where Paul is in the lead). Plot a histogram of both these quantities (using `plt.hist(data, density=True,bins='auto')`).


<hr>

### Ans.
Fill in answer

```python
# Imports
import random
import numpy as np
import matplotlib.pyplot as plt

n=1000
T=100
def sim(n, T):
  winnings=[]
  leading=[]

  S=0
  leadCount=0
  for i in range(n):
    S=0
    leadCount=0
    for j in range(T):
      S+= random.choice([-1, 1])
      if S>0:
        leadCount+=1

    winnings.append(S/T)
    leading.append(leadCount/T)




  plt.hist(winnings, density=True, bins='auto')
  plt.hist(leading, density=True, bins='auto')


sim(1000,100)
```

### Ans.


As T approaches infinity, we expect that the approximately normal distribution of $Z_T$ will tighten around 0.

We have that:

Z_T resembles a normal distribution as it is a sum of many IID binomial rvs

Each step is a rv $X_t$ with variance 1 (always 1 away from expectation of 0)

Then Var($S_T$)= $\sum_{t=1}^TVar(X_t)=T$

Then $Var(Z_T)=Var(S_T/T)=1/T^2*Var(S_T)=1/T$

Finally, we find the standard deviation as $\sqrt(1/T)=1/\sqrt(T)$

Therefore, the standard deviation of the normal distribution for $S_T$/T will decrease as T increases, creating a tighter normal distribution around 0.


Secondly, we expect $W_T$ to be a distribution centered around .5, but with most of the mass near 0 and 1. This makes sense because we expect random walks to take most of their steps far away from 0, meaning the amount of times that we cross from negative values to positive is limited, and predominantly in the early steps. **c.** Define $N_{circle}$ to be the (random) number of points which lie within the circle. What is $\mathbb{E}[N_{circle}]$ and $Var(N_{circle})$ (express your answer in terms of $\pi$)?

Suppose also that you did not know the value of $\pi$; can you still argue that $Var(N_{circle})\leq \frac{n}{4}$?


<hr>

### Ans.

The probability that a randomly chosen point will be within the circle is equal to the area of the circle over the area of the square.

So, $p = \frac{\pi \times 1^2}{2 \times 2} = \frac{\pi}{4}$

The distribution of $N_{circle}$ follows a Binomial(n,p) distribution where n = 10,000 and p = $\frac{\pi}{4}$

So, $E[N_{circle}] = n \times p = 2,500 \pi$

And $Var(N_{circle}) = n \times p \times (1-p) = 2,500\pi (\frac{4-\pi}{4})
$

If we did not know the value of pi, we could still argue that $Var(N_{circle})\leq \frac{n}{4}$ by finding the max possible variance of one trial.

$\frac{d}{dp}p\times (1-p)=0\\
1 - 2p = 0\\
p = \tfrac 12$

Now that we know what p is for the maximum possible variance in a single trial, plugging it into the binomial variance equation gives us:

$n\times \tfrac 12 \times \tfrac 12 = \tfrac n4$

And so we can be sure that $Var(N_{circle})$ will be less than or equal to $\tfrac n4$

**d.** Given $n$ random points generated as in part (b), we can estimate $\pi$ as $\pi_n = \frac{4N_{circle}}{n}$ for $\pi$ (where as before $N_{circle}$ is the random number of points that lie inside the circle). For $n$ between $1$ and $1000$, plot the relative error $|\pi_n - \pi|/\pi$. Also plot the function $1/\sqrt{n}$ on the same plot. How do the relative errors compare to this function?


<hr>

### Ans.

```python
errors = []

for n in range(1, 1001):
  X = []
  Y = []
  Z = []

  for i in range(n):
      X.append(np.random.uniform(-1.0, 1.0))
      Y.append(np.random.uniform(-1.0, 1.0))
      Z.append(1 if X[i]**2 + Y[i]**2 <= 1 else 0)

  pi_n = 4 * sum(Z) / n
  rel_er = abs(pi_n - np.pi) / np.pi
  errors.append(rel_er)

# Make plot perfectly square
plt.figure(figsize=(15, 8))
plt.scatter(range(1, 1001), errors, s=2.0, cmap=plt.cm.Paired)
plt.plot(range(1, 1001), [1/np.sqrt(n) for n in range(1, 1001)])
plt.show()
```

We can see that the relative errors closely follow the plotted function, decreasing as n increases.
